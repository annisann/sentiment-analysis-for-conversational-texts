## tf_idf.ipynb
code for creating tf idf value from all token all chats

```python
import pandas as pd
import math
```

### 1. read raw tf csv and..
count df {document freq.} for each token each chat

```python
raw_tf = pd.read_csv('raw_tf.csv')
```

```python
raw_tf.head()
```

```python
## count df
df = list()
for _, row in raw_tf.iterrows():
    _temp = row[1:] ## exclude token at index 0
    _temp = [i for i in _temp if i != 0]
    df_count = len(_temp)
    df.append(df_count)

# add df list to raw_tf column
raw_tf['df'] = df
```

```python
raw_tf.head()
```

### 2. count log freq.
1 + log10(tf) if tf > 0, 0 if tf = 0

```python
tf_idf = raw_tf.copy() ## deep copy raw_tf df
```

```python
tf_idf.head()
```

```python
## count log freq.
for i, row in tf_idf.iterrows():
    token = row[0]
    _df = row[-1] 
    _temp = row[1:-1].tolist()
    
    for j, value in enumerate(_temp):
        if value > 0:
            value = 1 + (math.log(value, 10))
            _temp[j] = value
    
    _temp.insert(0, token)
    _temp.append(_df)
    
    tf_idf.iloc[i] = _temp
```

```python
tf_idf.head()
```

### 3. count idf
log10 (N / df), N is total document

```python
N = tf_idf.shape[1] - 2 ## 2 is tokens and df columns

idf = [math.log((N / df_value), 10) for df_value in df]

tf_idf.drop('df', axis=1, inplace=True)
tf_idf['idf'] = idf
```

```python
tf_idf.head()
```

### 4. count tf-idf
log freq. * idf

```python
for i, row in tf_idf.iterrows():
    _temp = row[1:-1]
    token = row[0]
    _idf = row[-1] 
    
    _temp = _temp * idf[i]
    
    _temp = _temp.tolist()
    _temp.insert(0, token)
    _temp.append(_idf)
    
    tf_idf.iloc[i] = _temp
```

```python
tf_idf.drop('idf', axis=1, inplace=True) ## drop idf column so tf idf df has same format as raw tf df
```

```python
tf_idf.head()
```

### 5. export tf-idf result to csv

```python
tf_idf.to_csv('tf_idf.csv', index=False)
```
